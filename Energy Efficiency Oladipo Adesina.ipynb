```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('energydata_complete.csv')
raw_data.head()
```

```python
raw_data.reset_index(drop=True)
data = raw_data.set_index('date')
data.head()
```

### Question 12

```python
from sklearn.linear_model import LinearRegression
```

```python
x = data['T2'].values.reshape(-1,1)
y = data['T6'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(x,y)
np.round(reg.score(x,y), 2)
```

```python
data_quiz = raw_data.drop(['date', 'lights'], axis = 1)
data_quiz.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

x1 = data.iloc[:, 1:]
y1 = data['Appliances']
scaler = MinMaxScaler()

x_scaled = scaler.fit_transform(x1)
x_scaled
```

```python
from sklearn.model_selection import train_test_split as tts

x_train, x_test, y_train, y_test = tts(x_scaled, y1, test_size = 0.3, random_state = 42)

reg1 = LinearRegression()
reg1.fit(x_train, y_train)
reg1.score(x_train, y_train)
```

```python
y_pred = reg1.predict(x_test)
```

### Question 13

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
np.round(mae, 2)
```

### Question 14

```python
rss = np.sum(np.square(y_test-y_pred))
np.round(rss,2)
```

### Question 15

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
np.round(rmse,3)
```

### Question 16

```python
from sklearn.metrics import r2_score
coef_deter = r2_score(y_test, y_pred)
np.round(coef_deter, 2)
```

### Question 18

```python
from sklearn.linear_model import Ridge

reg_ridge = Ridge(alpha=0.4)
reg_ridge.fit(x_train, y_train)

y_pred_ridge = reg_ridge.predict(x_test)
```

```python
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
print(rmse_ridge)
print(rmse)
```
